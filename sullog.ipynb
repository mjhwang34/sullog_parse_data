```python
import csv
import json
```

```python
json_file_path = "C:\sullog\\alchol_data\\alchol.json"
csv_file_path = "C:\sullog\\alchol_data\\alchol.csv"
```

```python
with open(json_file_path, 'r', encoding='utf8') as file:
    json_data = json.load(file)
    
alchol_data_json_string = json.dumps(json_data, ensure_ascii=False)
alchol_data_json = json.loads(alchol_data_json_string)
#print(alchol_data_json['data'])
csv_header = ['name', 'type', 'abv', 'volume', 'price', 'ingredients', 'manufacturer', 'owner', 'lat', 'lng', 'province', 'city', 'phone']
csv_list = []
for data in alchol_data_json['data']:
    temp = []
    temp.append(data['name'])
    temp.append(data['type'])
    temp.append(data['ABV'])
    temp.append(data['volume'])
    temp.append(data['price'])
    ingredient_list = data['ingredients']
    ingredient_string = ",".join(ingredient_list)
    temp.append(ingredient_string)
    temp.append(data['manufacturer'])
    temp.append(data['owner'])
    temp.append(data['address']['lat'])
    temp.append(data['address']['lng'])
    temp.append(data['address']['province'])
    temp.append(data['address']['city'])
    temp.append(data['phone'])
    csv_list.append(temp)
print(csv_list)

check=True
with open(csv_file_path, 'w', encoding='utf8', newline='') as file:
        writer = csv.writer(file)
        if(check == True):
            check = False
            # write the header
            writer.writerow(csv_header)
        # write multiple rows
        writer.writerows(csv_list)  
```
